BÀI 1: Phân loại ảnh trên bộ dữ liệu CIFAR-10
Bài này sử dụng bộ dữ liệu CIFAR-10 gồm 10 lớp có sẵn trong TensorFlow.

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

# 1. Tải và tiền xử lý dữ liệu CIFAR-10
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0  # Chuẩn hóa về [0, 1]

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# 2. Xây dựng mô hình CNN
model_cifar10 = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 3. Biên dịch mô hình
model_cifar10.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# 4. Huấn luyện mô hình (Thử nghiệm 10 epochs)
history = model_cifar10.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model_cifar10.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# 5. Đánh giá và Lưu mô hình để dùng cho bài 5
test_loss, test_acc = model_cifar10.evaluate(X_test, y_test, verbose=2)
print(f"\nĐộ chính xác trên tập kiểm tra: {test_acc*100:.2f}%")
model_cifar10.save('model_cifar10.h5')
```

BÀI 2: Phân loại ảnh Cat hoặc Dog (Nhị phân)

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Giả lập dữ liệu Cat/Dog (Thay thế bằng tập dữ liệu của giảng viên)
# 1000 ảnh train kích thước 64x64x3, nhãn 0 (Cat) hoặc 1 (Dog)
X_train_cd = np.random.rand(1000, 64, 64, 3).astype('float32')
y_train_cd = np.random.randint(2, size=(1000, 1))
X_test_cd = np.random.rand(200, 64, 64, 3).astype('float32')
y_test_cd = np.random.randint(2, size=(200, 1))

# Xây dựng mô hình CNN phân loại nhị phân
model_cat_dog = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid') # Dùng sigmoid cho phân loại nhị phân
])

model_cat_dog.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

# Huấn luyện
model_cat_dog.fit(X_train_cd, y_train_cd, epochs=5, batch_size=32, validation_data=(X_test_cd, y_test_cd))
model_cat_dog.save('model_cat_dog.h5')
```

BÀI 3: Phân loại ảnh Fashion-MNIST

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# 1. Tải dữ liệu Fashion-MNIST có sẵn
(X_train_f, y_train_f), (X_test_f, y_test_f) = datasets.fashion_mnist.load_data()

# Chuẩn hóa và reshape vì ảnh Fashion-MNIST là ảnh xám (1 kênh màu)
X_train_f = X_train_f.reshape((X_train_f.shape[0], 28, 28, 1)) / 255.0
X_test_f = X_test_f.reshape((X_test_f.shape[0], 28, 28, 1)) / 255.0

# 2. Xây dựng mô hình CNN
model_fashion = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_fashion.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# 3. Huấn luyện
model_fashion.fit(X_train_f, y_train_f, epochs=5, validation_data=(X_test_f, y_test_f))
model_fashion.save('model_fashion.h5')
```

BÀI 4: Phân loại khuôn mặt {Nam, Nữ}

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Giả lập dữ liệu khuôn mặt Nam/Nữ (Thay bằng dữ liệu thật khi chạy)
X_train_gender = np.random.rand(800, 64, 64, 3).astype('float32')
y_train_gender = np.random.randint(2, size=(800, 1))
X_test_gender = np.random.rand(200, 64, 64, 3).astype('float32')
y_test_gender = np.random.randint(2, size=(200, 1))

# Xây dựng mô hình CNN nhận diện giới tính
model_gender = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model_gender.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

# Huấn luyện
model_gender.fit(X_train_gender, y_train_gender, epochs=5, validation_data=(X_test_gender, y_test_gender))
model_gender.save('model_gender.h5')
```

BÀI 5: Triển khai ứng dụng WEB sử dụng Flask trên Google Colab

```python
import os
from flask import Flask, render_template_string, request, jsonify
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
from pyngrok import ngrok

# --- CẤU HÌNH NGROK TẠI ĐÂY ---
# Thay 'TOKEN_CỦA_BẠN_Ở_ĐÂY' bằng token lấy từ https://dashboard.ngrok.com/get-started/your-authtoken
NGROK_AUTH_TOKEN = "TOKEN_CỦA_BẠN_Ở_ĐÂY"
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

app = Flask(__name__)

# Tải trước các model đã train từ câu 1 - 4
models_dict = {
    "cifar10": load_model('model_cifar10.h5'),
    "cat_dog": load_model('model_cat_dog.h5'),
    "fashion": load_model('model_fashion.h5'),
    "gender": load_model('model_gender.h5')
}

# Danh sách nhãn tương ứng cho từng bài toán
labels_dict = {
    "cifar10": ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'],
    "cat_dog": ['Cat (Mèo)', 'Dog (Chó)'],
    "fashion": ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'],
    "gender": ['Nam (Male)', 'Nữ (Female)']
}

# Giao diện HTML
html_index = """
<!DOCTYPE html>
<html>
<head>
    <title>Deep Learning CNN Classifier</title>
    <meta charset="utf-8">
    <style>
        body { font-family: Arial, sans-serif; margin: 40px; text-align: center; background-color: #f4f4f9; }
        .container { max-width: 500px; margin: auto; background: white; padding: 20px; border-radius: 10px; box-shadow: 0px 0px 10px #ccc; }
        select, input, button { width: 100%; padding: 10px; margin: 10px 0; border-radius: 5px; border: 1px solid #ccc; }
        button { background-color: #4CAF50; color: white; border: none; cursor: pointer; font-size: 16px;}
        #result { margin-top: 20px; font-weight: bold; color: #333; font-size: 18px; }
    </style>
</head>
<body>
<div class="container">
    <h2>Thực hành CNN - Dự đoán ảnh</h2>
    <form id="upload-form" method="post" enctype="multipart/form-data">
        <label>Chọn bài toán cần dự đoán:</label>
        <select name="model_type" id="model_type">
            <option value="cifar10">Câu 1: CIFAR-10</option>
            <option value="cat_dog">Câu 2: Cat/Dog</option>
            <option value="fashion">Câu 3: Fashion-MNIST</option>
            <option value="gender">Câu 4: Nam/Nữ</option>
        </select>
        <input type="file" name="file" accept="image/*" required>
        <button type="submit">Dự Đoán Kết Quả</button>
    </form>
    <div id="result"></div>
</div>
<script>
    document.getElementById('upload-form').onsubmit = async (e) => {
        e.preventDefault();
        const formData = new FormData(e.target);
        document.getElementById('result').innerText = "Đang xử lý...";
        const response = await fetch('/predict', { method: 'POST', body: formData });
        const data = await response.json();
        if(data.success) {
            document.getElementById('result').innerHTML = `Kết quả dự đoán: <span style="color:blue">${data.prediction}</span>`;
        } else {
            document.getElementById('result').innerText = "Lỗi: " + data.error;
        }
    };
</script>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(html_index)

@app.route('/predict', methods=['POST'])
def predict():
    try:
        model_type = request.form.get('model_type')
        img_file = request.files['file']
        if not img_file: return jsonify({'success': False, 'error': 'No file'})

        filepath = "temp_img.png"
        img_file.save(filepath)

        if model_type in ["cat_dog", "gender"]:
            target_size, color_mode = (64, 64), 'rgb'
        elif model_type == "cifar10":
            target_size, color_mode = (32, 32), 'rgb'
        elif model_type == "fashion":
            target_size, color_mode = (28, 28), 'grayscale'

        img = image.load_img(filepath, target_size=target_size, color_mode=color_mode)
        img_array = image.img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        curr_model = models_dict[model_type]
        preds = curr_model.predict(img_array)

        if model_type in ["cat_dog", "gender"]:
            class_idx = int(preds[0][0] > 0.5)
        else:
            class_idx = np.argmax(preds[0])

        result_label = labels_dict[model_type][class_idx]
        os.remove(filepath)
        return jsonify({'success': True, 'prediction': result_label})
    except Exception as e:
        return jsonify({'success': False, 'error': str(e)})

# Chạy ngrok
public_url = ngrok.connect(5000)
print(f"👉 TRUY CẬP ĐƯỜNG LINK WEB TẠI ĐÂY: {public_url}")
app.run(port=5000)
```
