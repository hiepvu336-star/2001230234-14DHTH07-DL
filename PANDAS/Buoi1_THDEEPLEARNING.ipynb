```python
orig_dim1 = int(input("Nhập kích thước chiều 1 của tensor gốc (ví dụ: 10) (Bài tập 9): "))
orig_dim2 = int(input("Nhập kích thước chiều 2 của tensor gốc (ví dụ: 10) (Bài tập 9): "))
orig_dim3 = int(input("Nhập kích thước chiều 3 của tensor gốc (ví dụ: 3) (Bài tập 9): "))

# Yêu cầu người dùng nhập kích thước cắt, nhưng đảm bảo không lớn hơn kích thước gốc
crop_dim1_input = int(input("Nhập kích thước chiều 1 của tensor cắt (phải <= " + str(orig_dim1) + ") (Bài tập 9): "))
crop_dim2_input = int(input("Nhập kích thước chiều 2 của tensor cắt (phải <= " + str(orig_dim2) + ") (Bài tập 9): "))
crop_dim3_input = int(input("Nhập kích thước chiều 3 của tensor cắt (phải <= " + str(orig_dim3) + ") (Bài tập 9): "))

# Đảm bảo kích thước cắt không vượt quá kích thước gốc
crop_dim1 = min(crop_dim1_input, orig_dim1)
crop_dim2 = min(crop_dim2_input, orig_dim2)
crop_dim3 = min(crop_dim3_input, orig_dim3)

random_image = tf.random.uniform([orig_dim1, orig_dim2, orig_dim3], minval=0, maxval=256, dtype=tf.int32)
cropped_image = tf.image.random_crop(random_image, size=[crop_dim1, crop_dim2, crop_dim3])
print("Tensor gốc kích thước (", orig_dim1, ", ", orig_dim2, ", ", orig_dim3, "):")
print(random_image.shape)
print("Tensor sau khi cắt ngẫu nhiên kích thước (", crop_dim1, ", ", crop_dim2, ", ", crop_dim3, "):")
print(cropped_image.shape)
```

## Bài tập 10: Tạo ngẫu nhiên hai tensor 1-chiều x, y có 5 phần tử. Trả về x + y nếu x < y, và x – y nếu x > y

```python
num_elements_10 = int(input("Nhập số phần tử cho hai tensor 1-chiều x, y (Bài tập 10): "))
min_val_10 = float(input("Nhập giá trị nhỏ nhất cho các phần tử (Bài tập 10): "))
max_val_10 = float(input("Nhập giá trị lớn nhất cho các phần tử (Bài tập 10): "))

x_ten = tf.random.uniform([num_elements_10], minval=min_val_10, maxval=max_val_10, dtype=tf.float32)
y_ten = tf.random.uniform([num_elements_10], minval=min_val_10, maxval=max_val_10, dtype=tf.float32)

result_ten = tf.where(tf.less(x_ten, y_ten), x_ten + y_ten, x_ten - y_ten)

print("Tensor x:", x_ten)
print("Tensor y:", y_ten)
print("Kết quả (x+y nếu x<y, x-y nếu x>y):")
print(result_ten)
```

## Bài tập 12: Viết hàm tạo 2 tensor ngẫu nhiên có kích thước x = (3, 2, 2), y = (3, 2, 2). Viết hàm tính cộng, trừ, nhân, chia sau đó chuyển các hàm này vào trong graph.

```python
@tf.function
def create_random_tensors():
    x_twelve = tf.random.uniform([3, 2, 2], minval=0, maxval=10, dtype=tf.float32)
    y_twelve = tf.random.uniform([3, 2, 2], minval=0, maxval=10, dtype=tf.float32)
    return x_twelve, y_twelve

@tf.function
def add_tensors(a, b):
    return a + b

@tf.function
def subtract_tensors(a, b):
    return a - b

@tf.function
def multiply_tensors(a, b):
    return a * b

@tf.function
def divide_tensors(a, b):
    # Sử dụng tf.math.divide_no_nan để tránh lỗi chia cho 0
    return tf.math.divide_no_nan(a, b)

x_twelve_val, y_twelve_val = create_random_tensors()

print("Tensor x:")
print(x_twelve_val)
print("Tensor y:")
print(y_twelve_val)

print("Cộng:")
print(add_tensors(x_twelve_val, y_twelve_val))

print("Trừ:")
print(subtract_tensors(x_twelve_val, y_twelve_val))

print("Nhân:")
print(multiply_tensors(x_twelve_val, y_twelve_val))

print("Chia:")
print(divide_tensors(x_twelve_val, y_twelve_val))
```

## Bài tập 13: Xây dựng model để tính tổng hai tensor x, y

```python
@tf.function
def sum_model(x_thirteen, y_thirteen):
    return x_thirteen + y_thirteen

# Ví dụ sử dụng model
num_elements_13 = int(input("Nhập số phần tử cho hai tensor x, y (Bài tập 13): "))
min_val_13 = float(input("Nhập giá trị nhỏ nhất cho các phần tử (Bài tập 13): "))
max_val_13 = float(input("Nhập giá trị lớn nhất cho các phần tử (Bài tập 13): "))

x_thirteen_val = tf.random.uniform([num_elements_13], minval=min_val_13, maxval=max_val_13, dtype=tf.float32)
y_thirteen_val = tf.random.uniform([num_elements_13], minval=min_val_13, maxval=max_val_13, dtype=tf.float32)

print("Tensor x:", x_thirteen_val)
print("Tensor y:", y_thirteen_val)
print("Tổng hai tensor:", sum_model(x_thirteen_val, y_thirteen_val))
```

## Bài tập 14: Xây dựng model để tính hiệu và tích hai tensor x, y

```python
@tf.function
def diff_and_prod_model(x_fourteen, y_fourteen):
    difference = x_fourteen - y_fourteen
    product = x_fourteen * y_fourteen
    return difference, product

# Ví dụ sử dụng model
num_elements_14 = int(input("Nhập số phần tử cho hai tensor x, y (Bài tập 14): "))
min_val_14 = float(input("Nhập giá trị nhỏ nhất cho các phần tử (Bài tập 14): "))
max_val_14 = float(input("Nhập giá trị lớn nhất cho các phần tử (Bài tập 14): "))

x_fourteen_val = tf.random.uniform([num_elements_14], minval=min_val_14, maxval=max_val_14, dtype=tf.float32)
y_fourteen_val = tf.random.uniform([num_elements_14], minval=min_val_14, maxval=max_val_14, dtype=tf.float32)

diff, prod = diff_and_prod_model(x_fourteen_val, y_fourteen_val)

print("Tensor x:", x_fourteen_val)
print("Tensor y:", y_fourteen_val)
print("Hiệu hai tensor:", diff)
print("Tích hai tensor:", prod)
```

## Bài tập 15: Xây dựng model để tính chu vi hình chữ nhật P = (a + b) * 2

```python
@tf.function
def rectangle_perimeter_model(a_fifteen, b_fifteen):
    return (a_fifteen + b_fifteen) * 2

# Ví dụ sử dụng model
length = tf.constant(float(input("Nhập chiều dài hình chữ nhật (Bài tập 15): ")), dtype=tf.float32)
width = tf.constant(float(input("Nhập chiều rộng hình chữ nhật (Bài tập 15): ")), dtype=tf.float32)

perimeter = rectangle_perimeter_model(length, width)
print("Chiều dài a:", length.numpy())
print("Chiều rộng b:", width.numpy())
print("Chu vi hình chữ nhật:", perimeter.numpy())
```

## Bài tập 16: Xây dựng model để tính diện tích toàn phần hình hộp chữ nhật

```python
@tf.function
def cuboid_surface_area_model(a_sixteen, b_sixteen, c_sixteen):
    s_xung_quanh = (a_sixteen + b_sixteen) * c_sixteen * 2
    s_toan_phan = (a_sixteen * b_sixteen) * 2 + s_xung_quanh
    return s_toan_phan, s_xung_quanh

# Ví dụ sử dụng model
length_c = tf.constant(float(input("Nhập chiều dài hình hộp chữ nhật (Bài tập 16): ")), dtype=tf.float32)
width_c = tf.constant(float(input("Nhập chiều rộng hình hộp chữ nhật (Bài tập 16): ")), dtype=tf.float32)
height_c = tf.constant(float(input("Nhập chiều cao hình hộp chữ nhật (Bài tập 16): ")), dtype=tf.float32)

total_area, lateral_area = cuboid_surface_area_model(length_c, width_c, height_c)

print("Chiều dài a:", length_c.numpy())
print("Chiều rộng b:", width_c.numpy())
print("Chiều cao c:", height_c.numpy())
print("Diện tích xung quanh hình hộp chữ nhật:", lateral_area.numpy())
print("Diện tích toàn phần hình hộp chữ nhật:", total_area.numpy())
```

## Bài tập 17: Xây dựng model chứng minh hai cạnh bên hình thang song song theo điều kiện

```python

@tf.function
def check_parallel_vectors(vec1, vec2, tolerance=1e-6):
    cross_product_2d = vec1[0] * vec2[1] - vec1[1] * vec2[0]
    is_parallel = tf.abs(cross_product_2d) < tolerance
    return is_parallel

# Ví dụ: Hai cạnh bên của hình thang (có thể không song song)
leg1_str = input("Nhập vector cạnh bên 1 dưới dạng 'x,y' (ví dụ: 1.0,2.0) (Bài tập 17): ")
leg1_coords = [float(x.strip()) for x in leg1_str.split(',')]
leg1 = tf.constant(leg1_coords, dtype=tf.float32)

leg2_str = input("Nhập vector cạnh bên 2 dưới dạng 'x,y' (ví dụ: 1.5,3.1) (Bài tập 17): ")
leg2_coords = [float(x.strip()) for x in leg2_str.split(',')]
leg2 = tf.constant(leg2_coords, dtype=tf.float32)

print("Vector cạnh bên 1:", leg1.numpy())
print("Vector cạnh bên 2:", leg2.numpy())
print("Hai cạnh bên có song song không?", check_parallel_vectors(leg1, leg2).numpy())
```

## Bài tập 18: Xây dựng model chứng minh hai cạnh đáy hình thang bằng nhau theo điều kiện

```python


@tf.function
def check_equal_bases_model(base1_length, base2_length, tolerance=1e-6):
    return tf.abs(base1_length - base2_length) < tolerance

# Ví dụ: Các cạnh đáy của hình thang thông thường
base_a_input = float(input("Nhập độ dài cạnh đáy 1 (Bài tập 18): "))
base_b_input = float(input("Nhập độ dài cạnh đáy 2 (Bài tập 18): "))

base_a = tf.constant(base_a_input, dtype=tf.float32)
base_b = tf.constant(base_b_input, dtype=tf.float32)

print("Độ dài cạnh đáy 1:", base_a.numpy())
print("Độ dài cạnh đáy 2:", base_b.numpy())
print("Hai cạnh đáy có bằng nhau không?", check_equal_bases_model(base_a, base_b).numpy())
```

## Bài tập 19: Xây dựng model chứng minh NA = NC theo mô tả với AM, MB, AN, NC là đầu vào

```python
# Yêu cầu "chứng minh NA = NC" khi AM, MB, AN, NC là đầu vào ngụ ý rằng chúng ta chỉ cần kiểm tra sự bằng nhau của AN và NC.
# Không có mối liên hệ hình học cụ thể nào được mô tả liên quan đến AM và MB để "chứng minh" NA = NC dựa trên chúng.
# Do đó, model sẽ đơn giản là một hàm kiểm tra sự bằng nhau của hai giá trị AN và NC.

@tf.function
def prove_na_equals_nc_model(AM, MB, AN, NC, tolerance=1e-6):
    is_equal = tf.abs(AN - NC) < tolerance
    return is_equal

# Ví dụ sử dụng model
am_val = tf.constant(float(input("Nhập giá trị AM (Bài tập 19): ")), dtype=tf.float32)
mb_val = tf.constant(float(input("Nhập giá trị MB (Bài tập 19): ")), dtype=tf.float32)
an_val = tf.constant(float(input("Nhập giá trị AN (Bài tập 19): ")), dtype=tf.float32)
nc_val = tf.constant(float(input("Nhập giá trị NC (Bài tập 19): ")), dtype=tf.float32)

print("Giá trị AM:", am_val.numpy())
print("Giá trị MB:", mb_val.numpy())
print("Giá trị AN:", an_val.numpy())
print("Giá trị NC:", nc_val.numpy())
print("NA có bằng NC không?", prove_na_equals_nc_model(am_val, mb_val, an_val, nc_val).numpy())
```
