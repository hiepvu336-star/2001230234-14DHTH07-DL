```python
import tensorflow as tf
import numpy as np

# 3.1. Tạo Tensor 0-chiều (scalar)
scalar_tensor = tf.constant(4)
print("Scalar Tensor:", scalar_tensor)

# Tạo với kiểu dữ liệu cụ thể
scalar_int16 = tf.constant(4, dtype=tf.int16)
scalar_float32 = tf.constant(4.0, dtype=tf.float32)
scalar_string = tf.constant("TensorFlow")

print("Scalar Int16:", scalar_int16)
print("Scalar Float32:", scalar_float32)
print("Scalar String:", scalar_string)
```

```python
# 3.2. Tạo Tensor 1-chiều (vector)
vector_tensor = tf.constant([2.0, 3.0, 4.0])
print("Vector Tensor (1D):", vector_tensor)
print("Shape:", vector_tensor.shape)
```

```python
# 3.3. Tạo Tensor 2-chiều (matrix)
matrix_tensor = tf.constant([[1, 2],
                             [3, 4],
                             [5, 6]])
print("Matrix Tensor (2D):\n", matrix_tensor.numpy())
print("Shape:", matrix_tensor.shape)
```

```python
# Bài 6: Tạo tensor [3, 2] ngẫu nhiên Normal (mean=0, stddev=2)
bai6 = tf.random.normal([3, 2], mean=0.0, stddev=2.0)
print("Bài 6:\n", bai6.numpy())

# Bài 7: Tạo tensor [3, 2] ngẫu nhiên Uniform trong khoảng [0, 2]
bai7 = tf.random.uniform([3, 2], minval=0, maxval=2)
print("\nBài 7:\n", bai7.numpy())

# Bài 8: Shuffle tensor X = [[1, 2], [3, 4], [5, 6]] theo chiều thứ nhất
X_8 = tf.constant([[1, 2], [3, 4], [5, 6]])
bai8 = tf.random.shuffle(X_8)
print("\nBài 8 (Shuffle):\n", bai8.numpy())

# Bài 9: Tạo tensor (10, 10, 3) và crop thành (5, 5, 3)
large_tensor = tf.random.uniform([10, 10, 3])
bai9 = tf.image.random_crop(large_tensor, size=[5, 5, 3])
print("\nBài 9 (Crop size):", bai9.shape)

# Bài 10: Hai tensor x, y (5 phần tử). Trả về x + y nếu x < y, ngược lại x - y
x = tf.random.uniform([5], minval=0, maxval=10)
y = tf.random.uniform([5], minval=0, maxval=10)
bai10 = tf.where(x < y, x + y, x - y)
print("\nBài 10:")
print("x: ", x.numpy())
print("y: ", y.numpy())
print("Kết quả: ", bai10.numpy())
```

```python
# Bài 11: x(3,2) [5,10], y(2,3) [15,20]. Giữ đường chéo x và x + y
x_11 = tf.random.uniform([3, 2], minval=5, maxval=10)
y_11 = tf.random.uniform([2, 3], minval=15, maxval=20)
# Vì x không phải ma trận vuông, ta lấy đường chéo của phần vuông lớn nhất
diag_x = tf.linalg.diag_part(x_11)
# Thực hiện cộng (TensorFlow sẽ broadcast hoặc cần reshape tùy mục đích, ở đây ta cộng x và chuyển vị của y)
result_11 = x_11 + tf.transpose(y_11)
print("Bài 11 - Kết quả x + y.T:\n", result_11.numpy())

# Bài 12: Hàm tính toán trong Graph
@tf.function
def basic_ops(x, y):
    return x + y, x - y, x * y, x / y

x_12 = tf.random.uniform([3, 2, 2])
y_12 = tf.random.uniform([3, 2, 2])
add, sub, mul, div = basic_ops(x_12, y_12)
print("\nBài 12: Đã thực hiện tính toán trong graph.")

# Bài 13 & 14: Model tính tổng, hiệu, tích
class BasicModel(tf.Module):
    @tf.function
    def calculate(self, x, y):
        return x + y, x - y, x * y

model = BasicModel()
res_sum, res_sub, res_mul = model.calculate(tf.constant(10), tf.constant(5))
print(f"\nBài 13 & 14 (x=10, y=5): Tổng={res_sum}, Hiệu={res_sub}, Tích={res_mul}")

# Bài 15: Model tính chu vi P = (a + b) * 2
@tf.function
def tinh_chu_vi(a, b):
    return (a + b) * 2

print("\nBài 15 - Chu vi (a=5, b=10):", tinh_chu_vi(tf.constant(5), tf.constant(10)).numpy())
```

```python
# Bài 16: Model tính diện tích toàn phần hình hộp chữ nhật
@tf.function
def tinh_dien_tich_toan_phan(a, b, c):
    s_xung_quanh = (a + b) * c * 2
    s_toan_phan = (a * b) * 2 + s_xung_quanh
    return s_xung_quanh, s_toan_phan

s_xq, s_tp = tinh_dien_tich_toan_phan(tf.constant(3.0), tf.constant(4.0), tf.constant(5.0))
print("Bài 16 - S xung quanh:", s_xq.numpy(), ", S toàn phần:", s_tp.numpy())

# Bài 17 & 18: Model chứng minh tính chất hình thang
# Giả sử điều kiện song song hoặc bằng nhau được biểu diễn qua vector/tensor
@tf.function
def kiem_tra_hinh_thang(day_tren, day_duoi, canh_ben1, canh_ben2):
    # Ví dụ logic: nếu các vector hướng song song (tích vô hướng hoặc tỉ lệ)
    is_parallel = tf.reduce_all(tf.equal(day_tren / day_duoi, day_tren[0] / day_duoi[0]))
    return is_parallel

# Bài 19: Model chứng minh trung điểm (NA = NC)
@tf.function
def chung_minh_trung_diem(AN, NC):
    # Nếu AN - NC = 0 thì N là trung điểm AC
    is_midpoint = tf.abs(AN - NC) < 1e-6
    return is_midpoint

```
